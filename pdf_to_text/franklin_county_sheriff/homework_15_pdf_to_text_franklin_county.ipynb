Notes on PDF to text conversion:

to convert PDFs to text/useable files:

PDFS WHERE TEXT IS SELECTABLE:

use tabula

IF THE PDF IS ACTUALLY A SCREEN SHOT AND THE TEXT IS NOT SELECTABLE:

CONVERT IT TO PNG:
in the terminal, once in the directory where you want it saved and where it's located, type: 

convert -density 150 -antialias "input_file_name.pdf" -resize 1024x -quality 100 "output_file_name.png"

this will convert a PDF to a series of enumerated images.

or, to convert a PDF to a single image: 

convert -density 150 -antialias "input_file_name.pdf" -append -resize 1024x -quality 100 "output_file_name.png"

notes from here: https://aleksandarjakovljevic.com/convert-pdf-images-using-imagemagick/

then, while still in the directory, run this:

tesseract "file_name.png" stdout

now we have a png file.
