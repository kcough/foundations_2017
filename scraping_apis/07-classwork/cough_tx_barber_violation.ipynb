## Scraping

### Loop through each `tr` and print each person's name

You'll get an error because the first one doesn't have a name. How do you make that not happen? I'm happy to help if you ask on the board.

```python
#tr = table row. they're all in span tags, so we can use that to search. we have to find the ones where the span tags
#aren't empty
table_row = doc.find_all('tr')
for each_element in table_row:
    if each_element.find('span') != None:
        print(each_element.find('span').string)
```

## Loop through each `tr`, printing each violation description

- TIP: What is the container tag name for it?
- TIP: You'll get an error even if you're ALMOST right - which row is causing the problem?

```python
#it's in the second tr and the second td.
#here's what we do to find the violation: doc.find_all('td')[2]
#now add it into the loop 
#NOTE THE : after the 1, which means it'll go through all of them. BE CAREFUL because we don't want the first tr 
#tr[0]. that's just the name, violation etc. headers
table_row = doc.find_all('tr')[1:]
for each_element in table_row:
    print(each_element.find_all('td')[2].string)
```

## Loop through each `tr`, printing the complaint number

- TIP: It should be the last piece of the fist `td`

```python
#it's the class = results_text inside the td tag.
table_row = doc.find_all('tr')[1:]
for each_element in table_row:
    print(each_element.find_all(class_='results_text')[5].string)
#why are some of these numbers different?
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Person's name
- Violation description
- Violation number

Create a new dictionary for each `tr` (except the header).

```python
#okay we have to start by making an empty list
tx_barber_violations = []

table_row = doc.find_all('tr')
for each_element in table_row:
    current = {}
    #now here's where we make an empty dictionary and then add things to it using the loops we had earlier.
    if each_element.find('span') != None:
        current['name'] = each_element.find('span').string
        current['violation'] = each_element.find_all('td')[2].string
        current['complaint_number'] = each_element.find_all(class_='results_text')[5].string
        tx_barber_violations.append(current)
    
        #here we're saying, append my dictionary (current) inside the list tx_barber_violations
```

### Save that to a CSV

```python
#we need pandas and then we can safe it as a df and convert it to a csv
import pandas as pd
df = pd.DataFrame(tx_barber_violations)
df.to_csv("tx_barber_violations.csv", index=False)
```
